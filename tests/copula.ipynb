```python
from scipy.stats import multivariate_normal as mvn
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from scipy.stats import stats
from scipy.stats import norm
```

```python
sample=mvn.rvs(cov=[[1,0.7],[0.7,1]],size=1000)
```

```python
plt.plot(sample[:,0], sample[:,1], '.');
plt.title("MVN");
```

```python
stats.pearsonr(sample[:,0],sample[:,1])
```

```python
plt.hist(sample[:,0]);
plt.hist(sample[:,1]);
plt.title('marginals');
```

```python
uniform_x = mvn.cdf(sample[:,0])
uniform_y = mvn.cdf(sample[:,1])
```

```python
plt.plot(uniform_x, uniform_y,'.');
plt.title('transformed');
```

```python
stats.pearsonr(uniform_x,uniform_y)
```

```python
plt.hist(uniform_x);
plt.hist(uniform_y);
plt.title('marginals transformed');
```

```python
marginal_normal_x = norm.ppf(uniform_x)
marginal_normal_y = norm.ppf(uniform_y)
```

```python
plt.plot(marginal_normal_x, marginal_normal_y,'.');
plt.title('marginal_normal');
```

```python
stats.pearsonr(marginal_normal_x,marginal_normal_y)
```

```python
marginal_exp_x = -np.log(1-uniform_x)/2
marginal_exp_y = -np.log(1-uniform_y)/2
```

```python
plt.plot(marginal_exp_x, marginal_exp_y,'.');
plt.title('marginal_exp');
```

```python
stats.pearsonr(marginal_exp_x,marginal_exp_y)
```

```python
plt.hist(marginal_exp_x);
plt.hist(marginal_exp_y);
plt.title('marginals exp');
```
